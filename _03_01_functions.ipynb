# 함수(Functions)

## 역함수

### 아래의 일차 함수의 경우를 보자.
- f(x) = x + 1

```python
import numpy
import matplotlib.pyplot as plt
import math

plt.axvline(x=0, color = 'r') # draw x = 0 axes
plt.axhline(y=0, color = 'r') # draw y = 0 axes

def f(x):
    return x - 1

x = [a for a in numpy.arange(-5, 5, 0.1)] # list comprehension
y = [f(a) for a in x]

plt.plot(x, y)
plt.show()
```

```python
plt.axvline(x=0, color = 'r') # draw x = 0 axes
plt.axhline(y=0, color = 'r') # draw y = 0 axes

def f(x):
    return x - 1

x = [a for a in numpy.arange(-5, 5, 0.1)] # list comprehension
y = [f(a) for a in x]
plt.plot(x, y)

def f_inv(x):
    return x + 1

x2 = [a for a in numpy.arange(-5, 5, 0.1)] # list comprehension
y2 = [f_inv(a) for a in x]
plt.plot(x2, y2)

plt.show()
```

위와 같이 역함수는 원 함수와 y=x 대해서 대칭이다.

### 일대일 대응 함수
- f(x) = $e^x$
- 위의 지수함수를 일대일 대응 함수라고 하려면, 정의역은 R(실수), 공역(또는 치역)은 양의 실수라고 정의해야한다.

```python
import numpy
import matplotlib.pyplot as plt
import math

plt.axvline(x=0, color = 'r') # draw x = 0 axes
plt.axhline(y=0, color = 'r') # draw y = 0 axes

def f(x):
    return math.exp(x)

x = [a for a in numpy.arange(-5, 5, 0.1)] # list comprehension
y = [f(a) for a in x]

plt.plot(x, y)
plt.show()
```

위의 지수함수 f(x) = $e^x$의 역함수은 아래의 로그 함수이다.
- f(x) = ln(x)
- 이 함수의 정의역은 양의 실수, 공역은 R

```python
import numpy
import matplotlib.pyplot as plt
import math

plt.axvline(x=0, color = 'r') # draw x = 0 axes
plt.axhline(y=0, color = 'r') # draw y = 0 axes
plt.xlim([0, 10]) # x축 범위 지정
plt.ylim([0, 10]) # y축 범휘 지정

# 원 함수
def f(x):
    return math.exp(x)

x = [a for a in numpy.arange(0.1, 10, 0.1)] # list comprehension
y = [f(a) for a in x]
plt.plot(x, y)

# 역 함수
def g(x):
    return math.log(x) # Python의 자연로그함수

x = [a for a in numpy.arange(0.1, 10, 0.1)] # list comprehension
y = [g(a) for a in x]
plt.plot(x, y)

plt.show()
```

### 일대일 대응 함수가 아닌 경우


f = x$^2$
위의 함수는 '일대일 함수'가 아니다. 따라서 역함수를 논할수 없다.

따라서, 위의 함수의 정의역과 치역을 양의 실수로 정의하면, '일대일 대응함수'이기 때문에 역함수를 구할 수 있다.

위의 함수의 역함수 f$^{-1}$은

f$^{-1}$ = sqrt(x)


f = x$^2$를 그려보자. 정의역을 양의 실수로 하자.

```python
import numpy
import matplotlib.pyplot as plt

plt.axvline(x=0, color = 'r') # draw x = 0 axes
plt.axhline(y=0, color = 'r') # draw y = 0 axes

def f(x):
    return (x*x)

x = [a for a in numpy.arange(0, 5, 0.1)] # list comprehension
y = [f(a) for a in x]

plt.plot(x, y)
plt.show()
```

역함수

f$^{-1}$ = sqrt(x)

를 그려보자.

```python
math.sqrt(4)
```

```python
import math

def g(x):
    return math.sqrt(x)

plt.axvline(x=0, color = 'r') # draw x = 0 axes
plt.axhline(y=0, color = 'r') # draw y = 0 axes

x = [a for a in numpy.arange(0, 25, 0.1)] # list comprehension
y = [g(a) for a in x]

plt.plot(x, y)
plt.show()
```

위의 그림을 통해서,
원래 함수의 x와 y가 바뀐것을 눈으로 파악하자.

## 합성 함수

```python
def g(x):
    return x + 1

def f(x):
    return x * x

def composite_function(a, b, x):
    return b(a(x))

print(composite_function(g, f, 1))
```

위의 합성 함수는 향후에 인공신경망을 배울때 많이 사용된다.
합성 함수의 의미와, 향후에 합성 함수의 미분이 사용된다.
